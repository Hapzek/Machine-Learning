```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('F:\\Machine Learning\\ML Udemy\\0_Course Notebooks\\05-Seaborn\\dm_office_sales.csv')
df.head()
```

```python
plt.figure(figsize = (6,6), dpi = 100
sns.rugplot(x = 'salary', data = df, height = 0.5)
```

```python
# DO NOT USE # DISTPLOT ---> DEPRECATED
sns.set (style = 'darkgrid')        # style = 'whitegrid' 'dark' 'ticks'  <--- HIGH LEVEL COMMAND, need to restart kernel to go back to default
sns.displot(data = df, x = 'salary', bins = 10, color = '#1c1c27', kde = True, rug = True) # <--- HISTOGRAM bins = number of rectangules
```

```python
# HISTPLOT

sns.histplot(data = df, x= 'salary')
```

```python
sns.kdeplot(data = df, x = 'salary')
```
