```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('F:\\Machine Learning\\ML Udemy\\0_Course Notebooks\\05-Seaborn\\dm_office_sales.csv')
df.head()
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize = (10,4), dpi = 200)
sns.countplot(data = df, x = 'level of education', hue = 'division', palette = 'Set2')
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(dpi = 100, figsize = (15,8))
sns.barplot( data = df, x = 'level of education', y = 'salary', estimator = np.mean, ci = 'sd', hue = 'division')  # Standart Deviation = sd
plt.legend(bbox_to_anchor = (1.05,1))
```
