```python
import matplotlib.pyplot as plt
```

```python
# %matplotline inline for outdated jupyter notebooks
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)      # ; for not showing the object and location of the image
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xlim(0,10)
plt.ylim(0,20)
plt.title('String Title')
plt.savefig('0_myfirstplot.jpeg')
```
