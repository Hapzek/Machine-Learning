```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sintax fig.add_axes([0,0,1,1])
# first set 0,0 = Lower left corner of axes. To start in the middle of the image we can 0.5, 0.5 in this case.
# second set 1,1 = (width, height) this means the relation ratio between the "size" of the axes if (1,1) we have a 1:1 ratio. ********

fig = plt.figure(figsize = (5,5))   # To create an figure 720 x 720
axes = fig.add_axes([0,0,1,1])
```

```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
fig = plt.figure()
# LARGE SET OF AXES
axes1 = fig.add_axes([0,0,1,1])  # same size as the canvas
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of four.')

# SMALL SET OF AXES
axes2 = fig.add_axes([0.08,0.5,0.33,0.33])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,80)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in')
```
