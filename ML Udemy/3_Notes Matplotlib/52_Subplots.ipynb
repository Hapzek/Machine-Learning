```python
 import numpy as np
 import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(x,y)
axes[0][0].set_title('TITLE 0,0')

axes[0][1].plot(a,b)
axes[0][1].set_title('TITLE 0,1')

axes[1][0].plot(x,y)
axes[1][0].set_title('TITLE 1,0')

axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE 1,1')


fig.suptitle('Figure Level Title')
plt.tight_layout()  # No graphics overlapping

fig.savefig('new_subplots.jpeg', bbox_inches = 'tight')

# fig.subplots_adjust(wspace = 0.9) <--- for weight spacing, we can do the same with height spacing hspace
```
