```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,3*x, label = 'lento')
ax.plot(1/2*x,3*x, label = 'r√°pido')

plt.xlabel('Tiempo (min)')
plt.ylabel('Temperatura (C)')
plt.title('Temperatura vs Tiempo')

ax.legend(loc = 'best')
plt.savefig('grafica.jpeg')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label = 'x vs x')
ax.plot(x,x**2, label = 'x vs x^2')

ax.legend(loc = (1.1,0.5))
```

```python
### PART 2

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# ls options: '--' , '-' , '-.', ';'
ax.plot(x,x, label = 'x vs x', color ='#1c1c27', lw = 1, marker = 'o', ms = 5)
ax.plot(x,x**2, label = 'x vs x^2', color = 'orange', linewidth = 2.5, ls = '--')

ax.legend(loc = (1.01,0.5))
```
