```python
# CONCATENATE

import numpy as np
import pandas as pd
```

```python
data_one = {'A': ['A0', 'A1', 'A2', 'A3'],'B': ['B0', 'B1', 'B2', 'B3']}
```

```python
data_two = {'C': ['C0', 'C1', 'C2', 'C3'], 'D': ['D0', 'D1', 'D2', 'D3']}
```

```python
one = pd.DataFrame(data_one)
```

```python
two = pd.DataFrame(data_two)
```

```python
one
```

```python
two
```

```python
pd.concat([one,two], axis = 1)
```

```python
# If we CONCATENATE with axis = 0 we are concatenating the rows, pandas will fill with NaN the missing data.

pd.concat([one,two], axis = 0)
```

```python
# To avoid dealing wiht the NaN values and to combining correctly:
# First, we rename the name of the columns of the second dataframe.
two.columns = one.columns
two
```

```python
# Now we can concatenate without dealing with the NaN problem

pd.concat([one,two], axis = 0)
```

```python
# We still have duplicated index numbers, the solution for this is to:

mydf = pd.concat([one,two], axis = 0)
mydf
```

```python
# 2nd step

mydf.index = range(len(mydf))
```

```python
# done!

mydf
```
