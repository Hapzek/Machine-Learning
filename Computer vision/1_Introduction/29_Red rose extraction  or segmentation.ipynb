```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread(r'F:\\Machine Learning\\Computer vision\\1_Introduction\\tulips.jpg')
```

```python
plt.imshow(img[:, :, ::-1]) # BGR TO RGB
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(hsv)
```

```python
lr = np.array([0,120,70])
ur = np.array([15, 255, 255])

mask1 = cv2.inRange(hsv, lr, ur)
```

```python
type(mask1)
```

```python
mask1.dtype
```

```python
mask1.shape
```

```python
mask1.max()
```

```python
mask1.min()
```

```python
lr2 = np.array([160,120,70])
ur2 = np.array([180, 255, 255])

mask2 = cv2.inRange(hsv, lr2, ur2)
```

```python
mask2.max()
```

```python
mask2.min()
```

```python
np.unique(mask2)
```

```python
mask = mask1 | mask2 
```

```python
np.unique(mask)
```

```python
res = cv2.bitwise_and (img, img, mask = mask)
```

```python
plt.figure(1)
plt.subplot(121)
plt.imshow(img[:,:,::-1])
plt.subplot(122)
plt.imshow(res[:,:,::-1])
plt.show()
```
